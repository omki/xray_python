### Before and after 

```python
import csv
import matplotlib.pyplot as plt
from scipy.misc import imread
import numpy as np
from PIL import Image

before_x = []
before_y = []
with open('xrd/CuOGPinCuCl2.xy','r') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        before_x.append(row[0])
        before_y.append(row[1])
    
after_x = []
after_y = []
with open('xrd/CuOCuCl2.xy','r') as f:
    reader = csv.reader(f, delimiter='\t')
    for row in reader:
        after_x.append(row[0])
        after_y.append(row[1])
```

```python
plt.close()

f = plt.figure()#plt.figure(figsize=(12,3))
ax1 = plt.gca()
ax2 = ax1.twinx()

ln1 = ax1.plot(before_x, before_y, 'k-', label = 'before annealing')
ln2 = ax2.plot(after_x, after_y, 'r-', label = 'after annealing')
ax1.set_yscale('log')
ax2.set_yscale('log')
ax1.set_xlabel(r'$2 \theta$')
ax1.set_xticks([], [])
ax1.set_ylabel('counts before annealing', color='k')
ax2.set_ylabel('counts after annealing', color='r')
#ax1.axis('off')
#ax2.axis('off')

lines = ln1 + ln2
labels = [l.get_label() for l in lines]
plt.legend(lines, labels, loc=0)

plt.show()
```
